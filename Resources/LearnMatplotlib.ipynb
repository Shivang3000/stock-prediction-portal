### *Title & Labels*

### *Bar Chart*

```python
months = ['Jan','Feb','Mar','Apr']
revenue = [90,115,120,130]
```

```python
plt.figure(figsize=(10, 6))
plt.bar(months, revenue);
plt.title('Monthly revenue bar chart');
plt.xlabel('Months');
plt.ylabel('Revenue');
```

### *Subplots & Legends*

```python
days = [1,2,3,4,5]
original_price = [10,15,25,40,60]
predicted_price = [9,13,28,39,58]
```

```python
plt.plot(days, original_price, 'b-o', label='Original Price')
plt.plot(days, predicted_price, 'r-o', label='Predicted Price')
plt.legend()
plt.title('Plot with legends');
plt.xlabel('Weekdays');
plt.ylabel('Price');
```

### *Subplots*

```python
# First subplot original price
plt.subplot(3, 1, 1)
plt.plot(days, original_price, 'b-o', label='Original Price')
plt.legend()
plt.title('Original price over time');
plt.xlabel('Weekdays');
plt.ylabel('Price');

# Second subplot predicted price
plt.subplot(3, 1, 2)
plt.plot(days, predicted_price, 'r-o', label='Predicted Price')
plt.legend()
plt.title('Predicted price over time');
plt.xlabel('Weekdays');
plt.ylabel('Price');

# Third subplot difference between original & predicted price
price_difference = [o - p for o, p in zip(original_price,predicted_price)]
plt.subplot(3, 1, 3)
plt.bar(days, price_difference, color='purple', label='Price Difference')
plt.title('Difference between Original and Predicted Price')
plt.xlabel('Days')
plt.ylabel('Price Difference')
plt.legend()

plt.tight_layout()
```

### Saving the figure

```python
plt.plot(days, original_price, 'b-o', label='Original Price')
plt.plot(days, predicted_price, 'r-o', label='Predicted Price')
plt.legend()
plt.title('Plot with legends');
plt.xlabel('Weekdays');
plt.ylabel('Price');
plt.savefig('example_prediction.png', dpi=300)
```
